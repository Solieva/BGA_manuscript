```python
import numpy as np
import matplotlib.pyplot as plt
from tabulate import tabulate
```

```python
# temperature range 
temp = np.array([10,20,30,40,50]) # C

# K_M data
Km_4M = np.array([1.88,1.58,1.43,1.18,1.19])[::-1] # this was ordered from 50C to 10C, the [::-1] reverses this order to be 10C to 50C
Km_3M = np.array([2.63,2.30,2.02,2.01,1.64])[::-1]
Km_2M = np.array([2.85,2.75,2.69,2.65,2.54])[::-1]
Km_1M = np.array([3.84,2.89,2.83,2.73,2.66])[::-1]

Km_4M_error = np.array([0.11,0.06,0.07,0.10,0.10])[::-1]
Km_3M_error = np.array([0.24,0.09,0.09,0.30,0.16])[::-1]
Km_2M_error = np.array([0.27,0.21,0.17,0.21,0.19])[::-1]
Km_1M_error = np.array([0.14,0.13,0.22,0.40,0.20])[::-1]

# k_cat
kcat_4M = np.array([323.63,170.51,74.03,23.12,8.79])[::-1]
kcat_3M = np.array([392.88,205.87,80.87,32.59,10.35])[::-1]
kcat_2M = np.array([345.12,187.72,78.87,30.48,10.89])[::-1]
kcat_1M = np.array([183.15,86.27,16.89,4.08,1.76])[::-1]

kcat_4M_error = np.array([16.32,5.71,3.07,1.42,0.54])[::-1]
kcat_3M_error = np.array([32.92,7.69,3.11,4.37,0.87])[::-1]
kcat_2M_error = np.array([30.35,13.36,4.64,2.28,0.75])[::-1]
kcat_1M_error = np.array([6.30,3.54,1.26,0.56,0.12])[::-1]

# k_cat/K_M
efficiency_4M =  np.array([171.88,108.2,51.64,19.51,7.4])[::-1]
efficiency_3M =  np.array([149.62,89.36,39.95,16.24,6.3])[::-1]
efficiency_2M =  np.array([120.99,68.18,29.31,11.51,4.29])[::-1]
efficiency_1M =  np.array([47.67,29.82,5.97,1.49,0.66])[::-1]

efficiency_4M_error =  np.array([13.09,5.62,3.38,1.98,0.75])[::-1]
efficiency_3M_error =  np.array([18.36,4.94,2.30,3.27,0.81])[::-1]
efficiency_2M_error =  np.array([15.50,7.08,2.52,1.26,0.43])[::-1]
efficiency_1M_error =  np.array([2.36 ,1.78,0.65,0.30,0.07])[::-1]
```

```python
table_1M = [['10C','20C','30C', '40C', '50C'],
         Km_1M, Km_1M_error, kcat_1M, kcat_1M_error, efficiency_1M, efficiency_1M_error]
print("1M data")
print(tabulate(table_1M, headers='firstrow',showindex=["K_M","K_M error","k_cat","k_cat error", "k_cat/K_M", "k_cat/K_M error"],tablefmt='fancy_grid'))


table_2M = [['10C','20C','30C', '40C', '50C'],
         Km_2M, Km_2M_error, kcat_2M, kcat_2M_error, efficiency_2M, efficiency_2M_error]
print("2M data")
print(tabulate(table_2M, headers='firstrow',showindex=["K_M","K_M error","k_cat","k_cat error", "k_cat/K_M", "k_cat/K_M error"],tablefmt='fancy_grid'))


table_3M = [['10C','20C','30C', '40C', '50C'],
         Km_3M, Km_3M_error, kcat_3M, kcat_3M_error, efficiency_3M, efficiency_3M_error]
print("3M data")
print(tabulate(table_3M, headers='firstrow',showindex=["K_M","K_M error","k_cat","k_cat error", "k_cat/K_M", "k_cat/K_M error"],tablefmt='fancy_grid'))



table_4M = [['10C','20C','30C', '40C', '50C'],
         Km_4M, Km_4M_error, kcat_4M, kcat_4M_error, efficiency_4M, efficiency_4M_error]
print("4M data")
print(tabulate(table_4M, headers='firstrow',showindex=["K_M","K_M error","k_cat","k_cat error", "k_cat/K_M", "k_cat/K_M error"],tablefmt='fancy_grid'))
```

```python
params = {'mathtext.default': 'regular' }          
plt.rcParams.update(params)

figure, ax = plt.subplots(3, figsize=[7,14])
temp = [10,20,30,40,50]
ax[2].errorbar(temp,Km_1M,yerr=Km_1M_error,label="1M KCl",color="C0", alpha=0.9)
ax[2].errorbar(temp,Km_2M,yerr=Km_2M_error,label="2M KCl",color="C2", alpha=0.9)
ax[2].errorbar(temp,Km_3M,yerr=Km_3M_error,label="3M KCl",color="C1", alpha=0.9)
ax[2].errorbar(temp,Km_4M,yerr=Km_4M_error,label="4M KCl",color="C3", alpha=0.9)
ax[2].set_ylabel("$K_{M}$ (mM ONPG)",fontsize = 16)
ax[2].set_xlabel("Temperature ($^\circ$C)",fontsize = 16)
ax[2].tick_params(labelsize=14)

ax[1].errorbar(temp,kcat_1M,yerr=kcat_1M_error,label="1M KCl",color="C0", alpha=0.9)
ax[1].errorbar(temp,kcat_2M,yerr=kcat_2M_error,label="2M KCl",color="C2", alpha=0.9)
ax[1].errorbar(temp,kcat_3M,yerr=kcat_3M_error,label="3M KCl",color="C1", alpha=0.9)
ax[1].errorbar(temp,kcat_4M,yerr=kcat_4M_error,label="4M KCl",color="C3", alpha=0.9)
ax[1].set_ylabel("$k_{cat}$ ($sec^{-1}$)",fontsize = 16)
ax[1].set_xlabel("Temperature ($^\circ$C)",fontsize = 16)
ax[1].tick_params(labelsize=14)

ax[0].errorbar(temp,efficiency_1M,yerr=efficiency_1M_error,label="1M KCl",color="C0", alpha=0.9)
ax[0].errorbar(temp,efficiency_2M,yerr=efficiency_2M_error,label="2M KCl",color="C2", alpha=0.9)
ax[0].errorbar(temp,efficiency_3M,yerr=efficiency_3M_error,label="3M KCl",color="C1", alpha=0.9)
ax[0].errorbar(temp,efficiency_4M,yerr=efficiency_4M_error,label="4M KCl",color="C3", alpha=0.9)
ax[0].set_ylabel("$k_{cat}$ / $K_{M}$ ($sec^{-1} * mM^{-1} ONPG$)",fontsize = 16)
ax[0].set_xlabel("Temperature ($^\circ$C)",fontsize = 16)
ax[0].tick_params(labelsize=14)
ax[2].legend(fontsize=14, loc='upper center', bbox_to_anchor=(0.5, -0.3), ncol=4)

ax[0].text(1,200,"(a)",fontsize=14)
ax[1].text(1,450,"(b)",fontsize=14)
ax[2].text(1,4.1,"(c)",fontsize=14)
figure.tight_layout()

figure.savefig("../Figure1.png")
```
