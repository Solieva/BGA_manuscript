```python
# setup the figure and axes
fig = plt.figure(figsize=(18, 6))
ax1 = fig.add_subplot(121, projection='3d')

# data
_xx = RMSF_values
_yy = SASA
_zz = kcat_values
x, y, z = _xx.ravel(), _yy.ravel(), _zz.ravel()
# bar height/width/depth
top = z
bottom = np.zeros_like(top) # to make the bars reach the bottom
width = 0.0005
depth = 0.1 # how thin to make the bars. Originally set to 1

# plot
alphas = [0.2,0.4,0.6,0.8,1]
# 1M: 
x_data =      x[0:5] # 1M
y_data =      y[0:5] # 1M
z_data =    top[0:5] # 1M
b_data = bottom[0:5] # 1M
x_datapoints = [x_data[0],x_data[1],x_data[2],x_data[3],x_data[4]]
y_datapoints = [y_data[0],y_data[1],y_data[2],y_data[3],y_data[4]]
z_datapoints = [z_data[0],z_data[1],z_data[2],z_data[3],z_data[4]]
b_datapoints = [b_data[0],b_data[1],b_data[2],b_data[3],b_data[4]]
for x_datapoints_, y_datapoints_, z_datapoints_,b_datapoints_,alphas_ in zip(x_datapoints,y_datapoints,z_datapoints,b_datapoints,alphas):
    ax1.bar3d(x_datapoints_, y_datapoints_, b_datapoints_, width, depth, z_datapoints_, shade=False, color="C0", alpha=alphas_)
ax1.plot3D(x_datapoints, y_datapoints, z_datapoints, color="C0", label= "1M KCl")

# 2M: 
x_data =      x[5:10] # 2M
y_data =      y[5:10] # 2M
z_data =    top[5:10] # 2M
b_data = bottom[5:10] # 2M
x_datapoints = [x_data[0],x_data[1],x_data[2],x_data[3],x_data[4]]
y_datapoints = [y_data[0],y_data[1],y_data[2],y_data[3],y_data[4]]
z_datapoints = [z_data[0],z_data[1],z_data[2],z_data[3],z_data[4]]
b_datapoints = [b_data[0],b_data[1],b_data[2],b_data[3],b_data[4]]
for x_datapoints_, y_datapoints_, z_datapoints_,b_datapoints_,alphas_ in zip(x_datapoints,y_datapoints,z_datapoints,b_datapoints,alphas):
    ax1.bar3d(x_datapoints_, y_datapoints_, b_datapoints_, width, depth, z_datapoints_, shade=False, color="C2", alpha=alphas_)
ax1.plot3D(x_datapoints, y_datapoints, z_datapoints, color="C2", label= "2M KCl")

# 3M: 
x_data =      x[10:15] # 3M
y_data =      y[10:15] # 3M
z_data =    top[10:15] # 3M
b_data = bottom[10:15] # 3M
x_datapoints = [x_data[0],x_data[1],x_data[2],x_data[3],x_data[4]]
y_datapoints = [y_data[0],y_data[1],y_data[2],y_data[3],y_data[4]]
z_datapoints = [z_data[0],z_data[1],z_data[2],z_data[3],z_data[4]]
b_datapoints = [b_data[0],b_data[1],b_data[2],b_data[3],b_data[4]]
for x_datapoints_, y_datapoints_, z_datapoints_,b_datapoints_,alphas_ in zip(x_datapoints,y_datapoints,z_datapoints,b_datapoints,alphas):
    ax1.bar3d(x_datapoints_, y_datapoints_, b_datapoints_, width, depth, z_datapoints_, shade=False, color="C1", alpha=alphas_)
ax1.plot3D(x_datapoints, y_datapoints, z_datapoints, color="C1", label= "3M KCl")

# 4M: 
x_data =      x[15:20] # 4M
y_data =      y[15:20] # 4M
z_data =    top[15:20] # 4M
b_data = bottom[15:20] # 4M
x_datapoints = [x_data[0],x_data[1],x_data[2],x_data[3],x_data[4]]
y_datapoints = [y_data[0],y_data[1],y_data[2],y_data[3],y_data[4]]
z_datapoints = [z_data[0],z_data[1],z_data[2],z_data[3],z_data[4]]
b_datapoints = [b_data[0],b_data[1],b_data[2],b_data[3],b_data[4]]
for x_datapoints_, y_datapoints_, z_datapoints_,b_datapoints_,alphas_ in zip(x_datapoints,y_datapoints,z_datapoints,b_datapoints,alphas):
    ax1.bar3d(x_datapoints_, y_datapoints_, b_datapoints_, width, depth, z_datapoints_, shade=False, color="C3", alpha=alphas_)
ax1.plot3D(x_datapoints, y_datapoints, z_datapoints, color="C3", label= "4M KCl")

#ax1.set_title('3D plot') # title
ax1.view_init(20, 50)
ax1.grid(True)
ax1.set_xlabel("Average RMSF per residue (nm)",       fontsize=16,labelpad=15)
ax1.set_ylabel("Total SASA $\mathregular{(nm)^{2}}$", fontsize=16,labelpad=15)
ax1.set_zlabel("$k_{cat}$ ($sec^{-1}$)",rotation=90,  fontsize=16,labelpad=10)
ax1.zaxis.set_rotate_label(False) 

ax1.tick_params(axis="both", labelsize=14)
ax1.set_xticks(np.arange(0.08, 0.1, 0.005))
ax1.set_yticks(np.arange(310, 330, 5))

ax1.legend(fontsize = 14, loc='center', bbox_to_anchor=(1.10, 0.5))
ax1.text2D(-0.094 ,0.065 ,"(c)",fontsize=14)

fig.tight_layout()
plt.show()
fig.savefig("../Figure5_c.png")
```
