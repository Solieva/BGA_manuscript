```python
import numpy as np
import matplotlib.pyplot as plt
import mdtraj as md
```

Load in the RMSF values for the surface and nonsurface residues

```python
surface = np.load("../../root_mean_square_fluctuation/average_RMSF_per_residue/rmsf_surface_avg_values.npy", allow_pickle=True)
nonsurface =np.load("../../root_mean_square_fluctuation/average_RMSF_per_residue/rmsf_nonsurface_avg_values.npy", allow_pickle=True)
```

```python
# split the rmsf_surface_avg list into molarities  (or lists of 5) so graphing is easier
rmsf_nonsurface_avg_1M = nonsurface[0:5]
rmsf_nonsurface_avg_2M = nonsurface[5:10]
rmsf_nonsurface_avg_3M = nonsurface[10:15]
rmsf_nonsurface_avg_4M = nonsurface[15:20]

rmsf_surface_avg_1M = surface[0:5]
rmsf_surface_avg_2M = surface[5:10]
rmsf_surface_avg_3M = surface[10:15]
rmsf_surface_avg_4M = surface[15:20]
```

## Graph

```python
figure, (ax1, ax2) = plt.subplots(1,2, figsize=[12,4])


y1 = rmsf_nonsurface_avg_1M
y2 = rmsf_nonsurface_avg_2M
y3 = rmsf_nonsurface_avg_3M
y4 = rmsf_nonsurface_avg_4M

barWidth = 0.18

n = np.array([0,5,10,15,20])
ax1.bar(n+0, y1, color ='C0', label ='1 M')
ax1.bar(n+1, y2, color ='C2', label ='2 M')
ax1.bar(n+2, y3, color ='C1',  label ='3 M')
ax1.bar(n+3, y4, color ='C3', label ='4 M')

ax1.set_xlabel('Temperature ($^\circ$C)', fontsize=16)
ax1.set_ylabel('Average RMSF (nm)', fontsize=16)
ax1.set_xticks(n+1.5)
ax1.set_xticklabels(["10","20","30","40","50"], fontsize=14)
ax1.set_ylim(0,0.12)
ax1.set_title("Non-surface Residues", fontsize=16)
ax1.tick_params(labelsize=14)



y1 = rmsf_surface_avg_1M
y2 = rmsf_surface_avg_2M
y3 = rmsf_surface_avg_3M
y4 = rmsf_surface_avg_4M

ax2.bar(n+0, y1,  color ='C0', label ='1M KCl')
ax2.bar(n+1, y2,  color ='C2', label ='2M KCl')
ax2.bar(n+2, y3,  color ='C1', label ='3M KCl')
ax2.bar(n+3, y4,  color ='C3', label ='4M KCl')

ax2.set_xlabel('Temperature ($^\circ$C)', fontsize=16)
ax2.set_xticks(n+1.5)
ax2.set_xticklabels(["10","20","30","40","50"], fontsize=14)
#ax2.set_ylabel('Average RMSF (nm)', fontsize=16)
ax2.set_ylim(0,0.12)
ax2.set_title("Surface Residues", fontsize=16)
ax2.legend(loc='center left', bbox_to_anchor=(1, 0.5), fontsize=16)
ax2.tick_params(labelsize=14)

ax2.text(-6,.127,"(b)",fontsize=14)
ax1.text(-6,.127,"(a)",fontsize=14)

figure.tight_layout()
figure.savefig("../Figure4.png")
```
