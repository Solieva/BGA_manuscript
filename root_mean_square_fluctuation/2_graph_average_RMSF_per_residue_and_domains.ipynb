```python
import numpy as np
import matplotlib.pyplot as plt
import mdtraj as md
```

Load in the RMSF per residue data

```python
rmsf_avg_per_res = np.load("average_RMSF_per_residue/average_RMSF_per_residue.npy",allow_pickle=True)
rmsf_sd_per_res = np.load("average_RMSF_per_residue/standard_dev_RMSF_per_residue.npy",allow_pickle=True)

rmsf_avg_per_res_1M = rmsf_avg_per_res[0:5]
rmsf_avg_per_res_2M = rmsf_avg_per_res[5:10]
rmsf_avg_per_res_3M = rmsf_avg_per_res[10:15]
rmsf_avg_per_res_4M = rmsf_avg_per_res[15:20]

rmsf_sd_per_res_1M = rmsf_sd_per_res[0:5]
rmsf_sd_per_res_2M = rmsf_sd_per_res[5:10]
rmsf_sd_per_res_3M = rmsf_sd_per_res[10:15]
rmsf_sd_per_res_4M = rmsf_sd_per_res[15:20]

rmsf_avg_per_res_10C =[rmsf_avg_per_res_1M[0],rmsf_avg_per_res_2M[0],rmsf_avg_per_res_3M[0],rmsf_avg_per_res_4M[0]]
rmsf_avg_per_res_50C =[rmsf_avg_per_res_1M[4],rmsf_avg_per_res_2M[4],rmsf_avg_per_res_3M[4],rmsf_avg_per_res_4M[4]]
rmsf_sd_per_res_10C = [rmsf_sd_per_res_1M[0],rmsf_sd_per_res_2M[0],rmsf_sd_per_res_3M[0],rmsf_sd_per_res_4M[0]]
rmsf_sd_per_res_50C = [rmsf_sd_per_res_1M[4],rmsf_sd_per_res_2M[4],rmsf_sd_per_res_3M[4],rmsf_sd_per_res_4M[4]]
```

Load in the average RMSF per domain data

```python
rmsf_avg_DomainA = np.load("average_RMSF_per_residue/average_RMSF_Domain_A.npy",allow_pickle=True)
rmsf_avg_DomainB = np.load("average_RMSF_per_residue/average_RMSF_Domain_B.npy",allow_pickle=True)
rmsf_avg_DomainC = np.load("average_RMSF_per_residue/average_RMSF_Domain_C.npy",allow_pickle=True)

rmsf_avg_DomainA_1M = rmsf_avg_DomainA[0:5]
rmsf_avg_DomainA_2M = rmsf_avg_DomainA[5:10]
rmsf_avg_DomainA_3M = rmsf_avg_DomainA[10:15]
rmsf_avg_DomainA_4M = rmsf_avg_DomainA[15:20]

rmsf_avg_DomainB_1M = rmsf_avg_DomainB[0:5]
rmsf_avg_DomainB_2M = rmsf_avg_DomainB[5:10]
rmsf_avg_DomainB_3M = rmsf_avg_DomainB[10:15]
rmsf_avg_DomainB_4M = rmsf_avg_DomainB[15:20]

rmsf_avg_DomainC_1M = rmsf_avg_DomainC[0:5]
rmsf_avg_DomainC_2M = rmsf_avg_DomainC[5:10]
rmsf_avg_DomainC_3M = rmsf_avg_DomainC[10:15]
rmsf_avg_DomainC_4M = rmsf_avg_DomainC[15:20]

rmsf_avg_DomainA_10C =[rmsf_avg_DomainA_1M[0],rmsf_avg_DomainA_2M[0],rmsf_avg_DomainA_3M[0],rmsf_avg_DomainA_4M[0]]
rmsf_avg_DomainA_50C =[rmsf_avg_DomainA_1M[4],rmsf_avg_DomainA_2M[4],rmsf_avg_DomainA_3M[4],rmsf_avg_DomainA_4M[4]]
rmsf_avg_DomainB_10C =[rmsf_avg_DomainB_1M[0],rmsf_avg_DomainB_2M[0],rmsf_avg_DomainB_3M[0],rmsf_avg_DomainB_4M[0]]
rmsf_avg_DomainB_50C =[rmsf_avg_DomainB_1M[4],rmsf_avg_DomainB_2M[4],rmsf_avg_DomainB_3M[4],rmsf_avg_DomainB_4M[4]]
rmsf_avg_DomainC_10C =[rmsf_avg_DomainC_1M[0],rmsf_avg_DomainC_2M[0],rmsf_avg_DomainC_3M[0],rmsf_avg_DomainC_4M[0]]
rmsf_avg_DomainC_50C =[rmsf_avg_DomainC_1M[4],rmsf_avg_DomainC_2M[4],rmsf_avg_DomainC_3M[4],rmsf_avg_DomainC_4M[4]]
```

```python
labels_smaller = 16
labels_small = 20
labels_large = 24
```

```python
fig, ax1 = plt.subplots(2,figsize=(15,12))
x = np.arange(1,701) 

# plot the average RMSF values and standard deviations for 1-4M:
ax1[0].fill_between(x, rmsf_avg_per_res_10C[0]+rmsf_sd_per_res_10C[0], rmsf_avg_per_res_10C[0]-rmsf_sd_per_res_10C[0], color= "C0", alpha=0.2)
ax1[0].plot(x,rmsf_avg_per_res_10C[0],'-', color= "C0", label="1 M")
ax1[0].fill_between(x, rmsf_avg_per_res_10C[1]+rmsf_sd_per_res_10C[1], rmsf_avg_per_res_10C[1]-rmsf_sd_per_res_10C[1], color= "C2", alpha=0.2)
ax1[0].plot(x,rmsf_avg_per_res_10C[1],'-', color= "C2", label="2 M")
ax1[0].fill_between(x, rmsf_avg_per_res_10C[2]+rmsf_sd_per_res_10C[2], rmsf_avg_per_res_10C[2]-rmsf_sd_per_res_10C[2], color= "C1", alpha=0.2)
ax1[0].plot(x,rmsf_avg_per_res_10C[2],'-', color= "C1", label="3 M")
ax1[0].fill_between(x, rmsf_avg_per_res_10C[3]+rmsf_sd_per_res_10C[3], rmsf_avg_per_res_10C[3]-rmsf_sd_per_res_10C[3], color= "C3", alpha=0.2)
ax1[0].plot(x,rmsf_avg_per_res_10C[3],'-', color= "C3", label="4 M")
ax1[0].set_xlabel('Residue Number',fontsize=labels_small)
ax1[0].set_ylabel('Average RMSF (nm)',fontsize=labels_small)
ax1[0].set_title('10$^\circ$C',fontsize = labels_large)
ax1[0].set_ylim(0,1)
ax1[0].tick_params(labelsize=labels_small)

# plot the 3 domains in different colors and the active site residues
ax1[0].axvspan(1,   390, 0, 0.95,alpha=0.2, color='gray') # domain A: 1 through 390
ax1[0].axvspan(391, 614, 0, 0.95,alpha=0.3, color='gray') # domain B: 391 through 614
ax1[0].axvspan(615, 701, 0, 0.95,alpha=0.4, color='gray') # domain C: 615 through 700
ax1[0].axvspan(485, 503, 0.85, 0.9, alpha=0.2, color='navy') # large loop in domain B
ax1[0].axvspan(519, 551, 0.85, 0.9, alpha=0.4, color='navy') # large loop in domain B
ax1[0].axvspan(555, 577, 0.85, 0.9, alpha=0.6, color='navy') # large loop in domain B
ax1[0].axvspan(103, 103, 0, 0.95, alpha=0.5, color='C0' )
ax1[0].axvspan(141, 141, 0, 0.95, alpha=0.5, color='C1' )
ax1[0].axvspan(142, 142, 0, 0.95, alpha=0.5, color='C2' )
ax1[0].axvspan(267, 267, 0, 0.95, alpha=0.5, color='C3' )
ax1[0].axvspan(312, 312, 0, 0.95, alpha=0.5, color='C4' )
ax1[0].axvspan(320, 320, 0, 0.95, alpha=0.5, color='C5' )
ax1[0].axvspan(360, 360, 0, 0.95, alpha=0.5, color='C6' )
ax1[0].axvspan(363, 363, 0, 0.95, alpha=0.5, color='C7' )

# missing loops: Electron density was missing for 32 residues in three regions (residues 532–543, 657–672, and 697–700) 
# from (https://www.mdpi.com/2076-2607/8/10/1594/htm)
ax1[0].axvspan(532, 543, 0.75, 0.8, alpha=0.5, color='deeppink')
ax1[0].axvspan(657, 672, 0.75, 0.8, alpha=0.5, color='deeppink')
ax1[0].axvspan(697, 700, 0.75, 0.8, alpha=0.5, color='deeppink')

# label the vertical liines
rects = [103,141,267,312,360]
rects2 = [142,320,363]
labels = [103,141,267,312,360]
labels2 = [142,320,363]
colors= ['C0','C1','C3','C4','C6']
colors2 = ['C2','C5','C7']
colors3 = ["navy","deeppink"]
labels3 = ["large loops", "residues missing in PDB 6LVW"]
locations3 = [0.905,0.805]
rects3 = [519+4, 605]
labels4 = ["Domain A","Domain B","Domain C"]
rects4 = [200,500,650]
alphas = [1,1,1]
for rect, label, color_ in zip(rects,labels,colors):
    height = 0.01
    ax1[0].text(rect-9, 0.905, label, color=color_,ha="center", va="bottom",fontsize=12, rotation=0)
for rect, label, color_ in zip(rects2,labels2,colors2):
    height = 0.01
    ax1[0].text(rect+11, 0.905, label, color=color_,ha="center", va="bottom",fontsize=12, rotation=0)
for rect, label, color_,location in zip(rects3,labels3,colors3,locations3):
    height = 0.01
    ax1[0].text(rect+10,location, label,color=color_,ha="center", va="bottom",fontsize=12, rotation=0)
for rect, label, alpha_ in zip(rects4,labels4,alphas):
    height = 0.01
    ax1[0].text(rect, 0.95, label, alpha=alpha_,color="k",ha="center", va="bottom",fontsize=12, rotation=0)

# bar chart inset graph 
left, bottom, width, height = [0.18, 0.76, 0.3, 0.15]
ax2 = fig.add_axes([left, bottom, width, height])
barWidth = 0.18
y1 = rmsf_avg_DomainA_10C
y2 = rmsf_avg_DomainB_10C
y3 = rmsf_avg_DomainC_10C
# Set position of bar on X axis
x1 = np.arange(4)
x2 = [x + barWidth for x in x1]
x3 = [x + barWidth for x in x2]
x4 = [x + barWidth for x in x3]
# Make the barplot
ax2.bar(x1, y1, alpha = 0.1,color ='gray', width = barWidth,edgecolor ='k', label ='Domain A')
ax2.bar(x2, y2, alpha = 0.3, color ='gray', width = barWidth, edgecolor ='k', label ='Domain B')
ax2.bar(x3, y3, alpha = 0.5, color ='gray', width = barWidth, edgecolor ='k', label ='Domain C')
# label each bar
rects = ax2.patches
labels = ["Domain A", "Domain A", "Domain A", "Domain A", "Domain B","Domain B", "Domain B", "Domain B","Domain C","Domain C","Domain C","Domain C"]
for rect, label in zip(rects, labels):
    height = rect.get_height()
    ax2.text(rect.get_x() + rect.get_width() / 2, 0.005, label, ha="center", va="bottom", rotation=90)
# Adding Xticks
ax2.set_xlabel('Molarity (M)',fontsize=labels_small)
ax2.set_ylabel('Average RMSF',fontsize=labels_small)
ax2.set_xticks([x2[0],x2[1],x2[2],x2[3]])
ax2.set_xticklabels([1,2,3,4])
ax2.tick_params(labelsize=labels_smaller)
ax2.set_ylim(0,0.17)
```
