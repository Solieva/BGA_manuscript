```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(15,7))

#runs = [5,6,7,13,14,15,21,22,23,26,27,30,31,34,35,38,39,45,46,47]
runs = [5,6,7,26,27,45,46,47,38,39,13,14,15,30,31,21,22,23,34,35] #fixed the order

labels = ["1M, 273K","1M, 283K", "1M, 293K", "1M, 303K",
         "1M, 273K","1M, 283K", "1M, 293K", "1M, 303K", 
         "3M, 273K","3M, 283K", "3M, 293K", "3M, 303K", 
         "3M, 273K","3M, 283K", "3M, 293K", "3M, 303K", 
         "4M, 273K","4M, 283K", "4M, 293K", "4M, 303K", 
         "4M, 273K","4M, 283K", "4M, 293K", "4M, 303K", 
          "1M, 313K","1M, 323K","1M, 313K","1M, 323K",
          "3M, 313K","3M, 323K","3M, 313K","3M, 323K",
          "4M, 313K","4M, 323K","4M, 313K","4M, 323K",
          "2M, 313K","2M, 323K","2M, 313K","2M, 323K",
         "2M, 273K","2M, 283K", "2M, 293K", "2M, 303K", 
         "2M, 273K","2M, 283K", "2M, 293K", "2M, 303K" ]
Total_ns = []
for i in runs:
    data = np.genfromtxt("traj_lengths/dist_of_curated_RUN%d.csv"%i, delimiter=",")
    data_ns = data/10
    Total_ns.append(sum(data_ns))
    y = np.arange(0,len(data_ns))
    l = labels[i]
    plt.plot(data_ns,y,label="%s"%l)
    plt.ylabel("number of trajectories", fontsize=16)
    plt.xlabel("length of trajectory (ns)", fontsize=16)
    plt.legend(fontsize = 14, loc='center left', bbox_to_anchor=(1, 0.5), ncol=3)
plt.title("Curated Dataset Trajectory Lengths for p16925 and p14188, per RUN", fontsize=16)
plt.tight_layout()
plt.savefig("trajectory_lengths.png")
```
